```python
import pandas as pd
```

```python
url = "https://gist.githubusercontent.com/tgcsantos/3bdb29eba6ce391e90df2b72205ba891/raw/22fa920e80c9fa209a9fccc8b52d74cc95d1599b/dados_imoveis.csv"
dados = pd.read_csv(url)
```

```python
dados
```

```python
imoveis = dados
```

```python
imoveis[["Moeda","Valor_Imovel","Tipo_Compra"]] = imoveis["Valor"].str.split(expand = True)
```

```python
imoveis
```

```python
imoveis_venda = imoveis[imoveis["Tipo_Compra"].isnull()]
```

```python
imoveis_venda
```

```python
imoveis_venda["Valor_Imovel"] = imoveis_venda["Valor_Imovel"].str.replace(".","").astype(float)
```

```python
imoveis_venda.info()
```

```python
imoveis_venda["Valor_m2"] = imoveis_venda["Valor_Imovel"]/imoveis_venda["Metragem"]
```

```python
imoveis_venda
```

```python
imoveis_por_bairro = imoveis_venda.groupby("Bairro").mean()
imoveis_por_bairro
```

```python
imoveis_por_bairro = imoveis_venda.groupby("Bairro").sum()
imoveis_por_bairro
```
